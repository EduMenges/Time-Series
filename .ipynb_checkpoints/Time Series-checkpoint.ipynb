*E AGORA PRA MOSTRAR O GRÁFICO NOVO?*

Nem fudendo que eu vou repetir código.

```python
def plotar(titulo: str, labelx: str, labely: str, x, y, dataset):
    """Plota um gráfico baseando-se nos estilos definidos."""
    sns.set_palette("Accent")
    sns.set_style("darkgrid")

    ax = sns.lineplot(x=x, y=y, data=dataset)
    ax.figure.set_size_inches(12, 6)
    ax.set_title(titulo, loc="left", fontsize=18)
    ax.set_xlabel(labelx, fontsize=14)
    ax.set_xlabel(labely, fontsize=14)
    return ax
```

```python
gp_aumento = plotar("Aumento das vendas da Alucar", "Tempo", "Aumento", alucar["mes"], alucar["aumento"], alucar)
```

## Decompondo a aceleração das vendas

```python
alucar["aceleracao"] = alucar["aumento"].diff()
alucar.head()
```

```python
gp_aceleracao = plotar("Aceleração das vendas da Alucar", "Tempo", "Aceleração", alucar["mes"], alucar["aceleracao"], alucar)
```

## Juntando os gráficos

```python
plt.figure(figsize=(16,12))

ax = plt.subplot(3,1,1)
ax.set_title("Análise da Alucar de 2017 e 2018", fontsize=18, loc="left")
sns.lineplot(x="mes", y="vendas", data=alucar)

plt.subplot(3,1,2)
sns.lineplot(x="mes", y="aumento", data=alucar)

plt.subplot(3,1,3)
sns.lineplot(x="mes", y="aceleracao", data=alucar)

ax
```

```python
def plot_comparacao(titulo: str, x: str, y, dataset):
    plt.figure(figsize=(16,12))

    ax = plt.subplot(3,1,1)
    ax.set_title(titulo, fontsize=18, loc="left")
    sns.lineplot(x=x, y=y[0], data=dataset)

    plt.subplot(3,1,2)
    sns.lineplot(x=x, y=y[1], data=dataset)

    plt.subplot(3,1,3)
    sns.lineplot(x=x, y=y[2], data=dataset)

    return ax
```

```python
plot_comparacao("Análise de Vendas da Alucar em 2017 e 2018", "mes", ("vendas", "aumento", "aceleracao"), alucar)
```

# Chocolura — Vendas diárias  

```python
vendas_por_dia = pd.read_csv("vendas_por_dia.csv", sep=",")
vendas_por_dia.head()
```

```python
print(f"Há {vendas_por_dia.shape[0]} registros nesta loja.")
print(f"Tem {vendas_por_dia.isnull().sum().sum()} valores nulos.")
```

```python
vendas_por_dia.dtypes
```

```python
vendas_por_dia["dia"] = pd.to_datetime(vendas_por_dia["dia"])
```

```python
vendas_por_dia.dtypes
```

```python
vendas_por_dia["aumento"] = vendas_por_dia.vendas.diff()
vendas_por_dia["aceleracao"] = vendas_por_dia.aumento.diff()
vendas_por_dia.head()
```

```python
plot_comparacao("Análise de vendas em Outubro e Novembro", "dia", ("vendas", "aumento", "aceleracao"), vendas_por_dia)
```

```python
vendas_por_dia["dia_da_semana"] = vendas_por_dia["dia"].dt.weekday
vendas_por_dia.head(8)
```

```python
dias_traduzidos = {
    0: "Segunda-feira",
    1: "Terça-feira",
    2: "Quarta-feira",
    3: "Quinta-feira",
    4: "Sexta-feira",
    5: "Sábado",
    6: "Domingo"
}

vendas_por_dia["dia_da_semana"] = vendas_por_dia["dia_da_semana"].map(dias_traduzidos)
```

```python
vendas_por_dia["dia_da_semana"].unique()
```

```python
vendas_por_dia.head()
```

```python
vendas_agrupadas = vendas_por_dia.groupby("dia_da_semana")["vendas", "aumento", "aceleracao"].mean()
```

```python
vendas_agrupadas
```

### Correlação das vendas diárias

```python
ax = plt.figure(figsize=(12,6))
ax.suptitle("Correlação das vendas diárias", fontsize=18, x=0.26, y=0.95)
autocorrelation_plot(vendas_por_dia["vendas"])
```

# Statsmodels

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
resultado = seasonal_decompose(chocolura["vendas"], period=3)
ax = resultado.plot()
```

```python
observacao = resultado.observed
tendencia = resultado.trend
sazonalidade = resultado.seasonal
ruido = resultado.resid
```

```python
resultado = pd.DataFrame({
    "observacao": observacao,
    "tendencia": tendencia,
    "sazonalidade": sazonalidade,
    "ruido": ruido
})

resultado.head()
```

```python
plot_comparacao("Exemplo de Statsmodels", resultado.index, ("observacao", "tendencia", "sazonalidade"), resultado)
```

```python
alucel["media_movel"] = alucel.vendas.rolling(7).mean()
alucel.tail()
```

```python
plotar("Análise de vendas com SMA de 7 dias",
      "Tempo", "Média móvel",
      "dia", "media_movel",
      alucel)
```

**Como será que fica com um período maior?**

```python
alucel["media_movel_21"] = alucel.vendas.rolling(21).mean()

plotar("Análise de vendas com SMA de 21 dias",
      "Tempo", "Média móvel",
      "dia", "media_movel_21",
      alucel)
```
